```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(open("/content/sample_data/sec-edgar-filings/LULU/10-K/0001397187-21-000009/filing-details.html"), "html.parser")
```

```python
text = soup.get_text()
```

```python
lines = (line.strip() for line in text.splitlines())
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
# drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)

print(text)
```

```python
!pip install spacy
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(text)
```

```python
for i, token in enumerate(doc.sents):
    #print('-->Sentence %d: %s' % (i, token.text))
    if 'Financial Highlights' in token.text:
       print('found on line %d: %s' % (i, token.text))
        
```
